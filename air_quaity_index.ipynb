```python
missing= pd.DataFrame(df.isna().sum()) #counting NaN in each column of cities dataframe and storing the count in a new dataframe
```

```python
df.fillna(0,inplace=True) #handling null values
```

```python
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date']) #changing 'Date' to date-time format
```

```python
df.info()
```

```python
df['month']=pd.DatetimeIndex(df['date']).month #making a new column by extracting month part of the Date column in cities dataframe
df['Year']=pd.DatetimeIndex(df['date']).year ##making a new column by extracting year part of the Date column in cities dataframe
```

Pandas DatetimeIndex.date attribute outputs an Index object containing the date values present in each of the entries of the DatetimeIndex object.

Now using this index we can easily data of particular year or particular range.

Syntax: DatetimeIndex.date

Return: numpy array of python datetime.date

```python
df.head()
```

Now, we'll merge columns of rspm & spm and make a new column named PM

Similarly, merging so2 & no2 and creating new column named Nitric

Similarly, merging Benzene, toluene,xylene and creating new column named BTX

```python
df['PM']=df['rspm']+df['spm']
df['Nitric']=df['so2']+df['no2']
```

```python
df.head()
```

```python
cities_group=df.groupby(['location','month','Year'])[['PM','Nitric','no2','so2']].mean()
```

```python
cities_group
```

```python
cities_group=cities_group.reset_index(['location','month','Year'])
```
